```python
# DataFrame filtrado
df_fil = df[df['agno'] == 2024]
```

```python
# Eliminar registros sin días trabajados
df_fil = df[df['dias_trabajados_anual'] > 0]
```

```python
df_fil.info()
```

```python
# Mapear variables

# Dependencia
map_depe = {1: "Municipal", 2:"Particular Subvencionado", 3: "Particular Pagado", 4: "Corporación Admin Delegada", 5: "SLEP" }

df_fil["depe"] = df_fil["cod_depe2"].map(map_depe)

# Urbano - Rural
df_fil["zona"] = df_fil["rural_rbd"].map({0: "Urbano", 1: "Rural"})

# Categoría asistencia
map_asistencia = {1: "Inasistencia Crítica", 2: "Inasistencia Grave", 3: "Inasistencia Reiterada", 4: "Asistencia esperada"}

df_fil["categoria_asistencia"] = df_fil["categoria_asis_anual"].map(map_asistencia)
```

```python
# Tabla agregación territorial

# Convertir la columna 'tasa_asistencia_anual' a numérica, manejando comas como decimales
df_fil['tasa_asistencia_anual'] = df_fil['tasa_asistencia_anual'].str.replace(',', '.', regex=False).astype(float)

# Por región
agg_region = (df_fil.groupby("nom_reg_rbd_a").agg(estudiantes=("mrun", "count"),
                                                  tasa_asistencia_promedio=("tasa_asistencia_anual", "mean")).reset_index())

agg_region.sort_values("tasa_asistencia_promedio").head()
```

```python
# Distribución de categorías de asistencia por región
dist_asistencia = (df_fil.groupby(['nom_reg_rbd_a', 'categoria_asistencia'])).size().reset_index(name='n_estudiantes')

# Porcentaje dentro de cada región
dist_asistencia["porcentaje"] = (dist_asistencia.groupby('nom_reg_rbd_a')['n_estudiantes'].transform(lambda x: x / x.sum()))
```

```python
# Tabla Dashboard Power BI

dashboard_data = (df_fil.groupby(['nom_reg_rbd_a', 'depe', 'zona', 'cod_ense2', 'categoria_asistencia'])
                  .agg(estudiantes=('mrun', 'count'), tasa_asistencia_promedio=('tasa_asistencia_anual', 'mean')).reset_index())

dashboard_data.head()
```

```python
# Exportar
dashboard_data.to_csv("asistencia_chile_2024_agg.csv", sep=",", decimal=".", index=False)

files.download("asistencia_chile_2024_agg.csv")
```
